# Research Question 

<li> What will most likely determine whether an adolescent is will be a smoker or not</li>

```python
#selecting predictors and target variables

predictors = data[['BIO_SEX', 'HISPANIC', 'WHITE', 'BLACK', 'NAMERICAN', 'ASIAN', 'age',
        'ALCEVR1', 'ALCPROBS1', 'marever1', 'cocever1', 'inhever1',
       'cigavail', 'DEP1', 'ESTEEM1', 'VIOL1', 'PASSIST', 'DEVIANT1',
       'SCHCONN1', 'GPA1', 'EXPEL1', 'FAMCONCT', 'PARACTV', 'PARPRES']]

target = data.TREG1
```

```python
#training and test sets

from sklearn.model_selection import train_test_split

pred_train, pred_test, tar_train, tar_test  = train_test_split(predictors, target, test_size=.3)
```

```python
#the shape of the split sets
print (pred_train.shape)
print (pred_test.shape)
print (tar_train.shape)
print (tar_test.shape)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

classifier = RandomForestClassifier(n_estimators = 25)
classifier = classifier.fit(pred_train, tar_train)
predictions = classifier.predict(pred_test)
```

```python
print("Confusion matrix: ")
print(confusion_matrix(tar_test, predictions))
print()
print("Accuracy: ", accuracy_score(tar_test, predictions))

```

```python
important_features = pd.Series(data = classifier.feature_importances_, index = predictors.columns)
important_features.sort_values(ascending = False, inplace = True)
print(important_features)
```

```python
trees = range(25)
accuracy = np.zeros(25)
for idx in range(len(trees)):
    classifier = RandomForestClassifier(n_estimators = idx + 1)
    classifier = classifier.fit(pred_train, tar_train)
    predictions = classifier.predict(pred_test)
    accuracy[idx] = accuracy_score(tar_test, predictions)
```

```python
plt.cla()
plt.plot(trees, accuracy)
plt.show()
```

# Findings

 The top 5 features that determine whether one is likely to be a smoker are:
 
 * Marijuana usage
 * Deviant behaviour
 * Age
 * School connectedness
 * GPA score
 
 these features can be used to predict with 84 percent accuracy.
